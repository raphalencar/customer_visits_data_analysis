Qual a media de visitas por dia e hora a cada categoria?

```python
# media de visitas por hora e dia a cada categoria
for cat in data['categoria'].unique():
    print("Categoria - {}: ".format(cat))
    for day in data['day_of_week'].unique():
        cat_day = data[(data['day_of_week'] == day) & (data['categoria'] == cat)]
        mean_cat_hour_day = cat_day.groupby(['categoria', 'day_of_week','hour'])['user'].count().mean()
        print("{}: {:.2f} visitas por hora.".format(day, mean_cat_hour_day))
    print('\n')
```

Dividir o dia em turnos: Madrugada / Manh / Tarde / Noite. A partir dos dados, gerar novos dados agregando eles.

```python
# criar coluna turno de acordo com a hora:  0 - madrugada; 1 - manh; 2 - tarde; 3 - noite;
def create_shift(hour):
    if (hour >= 0) and (hour < 5):
        return 0
    elif (hour >= 5) and (hour < 12):
        return 1
    elif (hour >= 12) and (hour < 17):
        return 2
    else:
        return 3

data['turno'] = data.apply(lambda x: create_shift(x['hour']), axis=1)
```

Em que localidades devem ser abertos novos restaurantes de acordo com o dia e horas que eles sao visitados?

Para isso podemos utilizar o modelo de **clusterizao KMeans** que funciona bem com latitude e longitude. Primeiro vamos excluir as dimenses que no fazem sentido para clusterizao. 

```python
data_cp = data.copy()

# trocando as strings por codigo
data_cp['categoria'] = pd.Categorical(data_cp['categoria'])
data_cp['categoria'] = data_cp['categoria'].cat.codes

data_cp['date'] = pd.to_datetime(data_cp['date'])
data_cp['day_of_week'] = data_cp['date'].dt.weekday

data_cp = data_cp.drop(['user', 'date', 'hour'], axis=1)

data_cp.head()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

random_state = 42

# com a converso, a categoria restaurante ficou com o cdigo 74
restaurant = data_cp[data_cp['categoria'] == 74]

for i in range(2, 41):
    clusterer = KMeans(n_clusters=i, random_state=random_state)
    clusterer.fit(restaurant)
    
    preds = clusterer.predict(restaurant)
    
    centers = clusterer.cluster_centers_
    
    score = silhouette_score(restaurant, clusterer.labels_, random_state=random_state)
    print("Para k = {0} o score : {1}".format(i, score))
```

```python
k = 3

model = KMeans(n_clusters=k, random_state=random_state)
model.fit(restaurant)

preds = model.predict(restaurant)
labels = model.labels_
centers = model.cluster_centers_

score = silhouette_score(restaurant, labels, random_state=random_state)
print("Para k = {0} o score : {1}".format(k, score))
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)
scatter = ax.scatter(restaurant['longitude'], restaurant['latitude'], c=preds, s=50)
plt.colorbar(scatter)
plt.show()
```
